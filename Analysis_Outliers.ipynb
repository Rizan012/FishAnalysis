```python
plt.figure(figsize=(20,10))
sns.barplot(x="species",y="length",data=data)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x="species",y="weight",data=data)
```

```python
for i in ["length","weight","w_l_ratio"]:
    sns.kdeplot(x=i ,data = data ,color="purple")
    plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.kdeplot(x="w_l_ratio",data=data,hue="species")
```

```python
for x in data.species.unique():
    plt.title(f"Length box plot {x}")
    sns.boxplot(y='weight', data=data[data["species"]==x], color="red")
    plt.show()
```

```python
for x in data.species.unique():
    plt.title(f"Length box plot {x}")
    sns.boxplot(y='length', data=data[data["species"]==x], color="green")
    plt.show()
```

```python
for x in data.species.unique():
    plt.title(f"Length box plot {x}")
    sns.boxplot(y='w_l_ratio', data=data[data["species"]==x], color="cyan")
    plt.show()
```
